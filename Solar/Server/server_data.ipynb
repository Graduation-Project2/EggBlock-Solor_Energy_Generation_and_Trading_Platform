```python
import requests
import csv
import pandas as pd

def upload():
    try:
        line = ""
        f = open("pre_value.csv", 'r')
        for tmp in f:
            line += tmp
            print(line)
        f.close()
        data = {'data': line}
        myUrl = 'http://wiffy.io/sen/upload4.php'
        requests.post(myUrl, data=data).text
    except:
        print('ERROR')

def download():
    try:
        myUrl = 'http://wiffy.io/sen/data.csv'
        res = requests.post(myUrl).text
        print(res)
        sres = res.split("\n")
        f = open("data.csv", 'w')
        for tmp in sres:
            f.write(tmp + "\n")
        f.close()
    except:
        print('ERROR')

def delete():
    try:
        myUrl = 'http://wiffy.io/sen/delete.php'
        requests.post(myUrl).text
    except:
        print('ERROR')
upload()
#download()
#df = pd.read_csv("data.csv")
#df
```
